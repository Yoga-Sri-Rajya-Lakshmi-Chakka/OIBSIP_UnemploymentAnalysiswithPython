**Data Preprocessing**

**Data** **Cleaning**

```python
f.isnull().sum()
```

```python
f.isna().sum()
```

```python
f = f.fillna(f.mean())
```

```python
f.columns = f.columns.str.strip()
print(f.columns)
```

```python
f = f.drop_duplicates(['Date', 'Region'])
```

```python
f= f.drop_duplicates()
```

```python
f['Date'] = pd.to_datetime(f['Date'])
```

```python
f.set_index('Date', inplace=True)
```

```python
f.shape
```

**Data reshaping**

```python
f.rename(columns={'Region': 'State'}, inplace=True)
```

```python
average_unemployment_rate = f.groupby('State')['Estimated Unemployment Rate (%)'].mean()
```

```python
state_with_highest_unemployment = average_unemployment_rate.idxmax()
print("State with the highest unemployment rate:", state_with_highest_unemployment)
highest_unemployment_rate = average_unemployment_rate.max()
print("Highest unemployment rate:", highest_unemployment_rate)
```

```python
state_with_lowest_unemployment = average_unemployment_rate.idxmin()
print("State with the lowest unemployment rate:", state_with_lowest_unemployment)
lowest_unemployment_rate = average_unemployment_rate.min()
print("Lowest unemployment rate:", lowest_unemployment_rate)
```

**Data Visualization**

```python
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=f)
plt.xticks(rotation=45)
plt.title('Unemployment Rate Over Time')
plt.show()
```

```python
state_avg_unemployment = f.groupby('State')['Estimated Unemployment Rate (%)'].mean().reset_index()
sns.barplot(x='State', y='Estimated Unemployment Rate (%)', data=state_avg_unemployment)
plt.xticks(rotation=90)
plt.title('Average Unemployment Rate by State')
plt.show()
```

```python
sns.scatterplot(x='Estimated Unemployment Rate (%)', y='Estimated Labour Participation Rate (%)', data=f)
plt.title('Unemployment Rate vs. Labor Participation Rate')
plt.show()
```

```python
sns.histplot(f['Estimated Unemployment Rate (%)'], bins=20, kde=True)
plt.xlabel('Estimated Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.title('Unemployment Rate Distribution')
plt.show()
```

```python
regions = ['South', 'Northeast', 'East','North','West']
filtered= f[f['Region.1'].isin(regions)]
sns.set_palette("husl")
sns.set_style("darkgrid")
plt.figure(figsize=(12, 6))

sns.lineplot(data=filtered, x=filtered.index, y='Estimated Labour Participation Rate (%)', hue='Region.1', estimator=None, lw=1.5, alpha=0.5, sort=False)

# Customize the plot
plt.title('Labor Participation Rate Over Time by Region')
plt.xlabel('Date')
plt.ylabel('Labor Participation Rate (%)')
plt.legend(title='Region', loc='upper right')

# Display the plot
plt.show()
```

```python
correlation_matrix = f.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
freq_state_unemployment =f.groupby(['Frequency', 'State'])['Estimated Unemployment Rate (%)'].mean().unstack()
freq_state_unemployment.plot(kind='bar', stacked=True)
plt.xlabel('State')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate by Frequency and State')
plt.legend(title='Frequency')
plt.show()
```

```python
sns.boxplot(x='State', y='Estimated Unemployment Rate (%)', data=f)
plt.xticks(rotation=90)
plt.title('Unemployment Rate by State')
plt.show()
```

```python
import plotly.express as px
sunburst_data = f.groupby(['Region.1','State'])['Estimated Employed'].sum().reset_index()
fig = px.sunburst(sunburst_data, path=['Region.1','State'], values='Estimated Employed')
fig.show()
```
